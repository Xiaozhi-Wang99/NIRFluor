```python
import numpy as np
import pandas as pd
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
data_path = './dataset/classifier_train_test/'
x_train = pd.read_csv(data_path + 'X_train.csv')
y_train = pd.read_csv(data_path + 'y_train.csv').values.ravel()

print(x_train.shape)
print(y_train.shape)
```

```python
tsne = TSNE(n_components=2, perplexity=50, learning_rate=100, random_state = 0)
train_tsne = tsne.fit_transform(x_train)
print(train_tsne.shape)

output_path = './Result/classifier/'
np.savetxt(output_path + 't-SNE_result.csv', train_tsne, delimiter=',')
```

```python
plt.figure(figsize=(6, 4))

#non-NIR
plt.scatter(train_tsne[y_train == 0, 0], train_tsne[y_train == 0, 1],
            c= [(0.98,0.76,0.72)], label='Train_non_NIR', alpha=0.5)
#NIR
plt.scatter(train_tsne[y_train == 1, 0], train_tsne[y_train == 1, 1],
            c=[(0.50,0.71,0.84)], label='Train_NIR', alpha=0.5)

plt.xlabel('t-SNE1')  
plt.ylabel('t-SNE2')
plt.title('t-SNE Visualization for Binary Classification')
plt.legend()
plt.savefig(output_path +'t-SNE_classification_visualization.png', format='png')
plt.show()
```
